```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import pixell
```

```python
# bin average data
ellmin = 10
ellmax = 100
ells = np.arange(ellmin, ellmax+1, 1)
ps = np.random.rand(ellmax-ellmin+1)

bin_width = 10
bins = np.arange(ellmin, ellmax, bin_width)
print(bins)
#bins = numpy.linspace(0,1,10)
# plt.plot(data)
# plt.show()
# plt.close()


binned_ps, bin_edges, binnumber = stats.binned_statistic(ells, ps, statistic='mean',bins=bins)
bin_center = (bin_edges[1:]+bin_edges[:-1])/2
# [1:] is for the 1 to last, [:-1] is for 0 to the second last
print('binned_ps:', binned_ps)
print('binn_edegs:', bin_edges)
print('binn_center:',bin_center)
plt.plot(binned_ps)
plt.show()
plt.close()
```

```python
from scipy import interpolate
new_ells = np.arange(ellmin+5, ellmax-15, 1)
smooth_ps = interpolate.interp1d(bin_center, binned_ps, bounds_error=False)(new_ells)
plt.plot(ells, ps, 'o', new_ells, smooth_ps, '-')
# doesn't smooth ps???
```

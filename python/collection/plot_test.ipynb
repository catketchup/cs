# object oriented plot method,  see Christine's tutorial

```python
ls = np.arange(1,10,0.1)

#fig, ax = plt.subplots(2,2,figsize=(15,12),sharex=True, sharey=True)
fig, ax = plt.subplots(2,2,figsize=(15,12))
ax[0,0].plot(ls, np.sin(ls))
ax[0,0].set_title('sin',color='green')
ax[0,0].set_xlabel('x')
ax[0,0].set_ylabel('y')
ax[0,0].legend(['sin'])
ax[0,0].text(1,0.25,r'sin plot',fontsize=14)

ax[0,1].plot(ls, np.cos(ls))
ax[1,0].plot(ls, np.tan(ls))
plt.show()
```

```python
# guass_beam
ells = np.arange(1,1000,1)
beam = np.exp(-(ells**2))
#plt.plot(ells, beam)
plt.plot(ells, 1/beam)
```

```python
# test subplot
x = np.arange(1, 10, 0.1)

fig = plt.figure(figsize=(6,5))
plt.subplot(3,1,1)
plt.plot(x, np.sin(x))

plt.subplot(3,1,2)
plt.plot(x, np.cos(x))

plt.subplot(3,1,3)
plt.plot(x, np.tan(x))

fig.tight_layout()
```

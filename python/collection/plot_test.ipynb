```python
import numpy as np
import matplotlib.pylab as plt
```

```python
ls = np.arange(1,10,0.1)
fig, ax = plt.subplots(2,2,figsize=(15,12))
ax[0,0].plot(ls, np.sin(ls))
ax[0,0].set_title('sin')
ax[0,0].set_xlabel('x')
ax[0,0].set_ylabel('y')
ax[0,0].legend(['sin'])

ax[0,1].plot(ls, np.cos(ls))
ax[1,0].plot(ls, np.tan(ls))

```
